### Step 3. Assign it to a variable apple

```python
df_apple
```

### Step 4.  Check out the type of the columns

```python
df_apple.dtypes
```

### Step 5. Transform the Date column as a datetime type

```python
df_apple.Date = df_apple.Date.astype('datetime64')
```

```python
df_apple.dtypes
```

### Step 6.  Set the date as the index

```python
#df_apple = df_apple.set_index(df_apple.Date)
```

```python
df_apple
```

### Step 7.  Is there any duplicate dates?

```python
df_apple.index.has_duplicates
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
df_apple = df_apple.sort_values('Date')
```

```python
df_apple = df_apple.set_index(df_apple.Date)
```

```python
df_apple
```

### Step 9. Get the last business day of each month

```python
df_apple['BMonthEnd'] = (df_apple.index + pd.offsets.BMonthEnd(0)).day
```

```python
df_apple
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
df_apple.index.max()-df_apple.index.min()
```

### Step 11.  How many months in the data we have?

```python
len(df_apple.groupby([df_apple.Date.dt.year, df_apple.Date.dt.month]).size())
```
